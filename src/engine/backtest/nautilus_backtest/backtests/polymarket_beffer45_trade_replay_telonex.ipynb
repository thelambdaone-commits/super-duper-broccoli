# @beffer45 Telonex Trade Replay Runner

Executes the hard-coded @beffer45 Telonex account-ledger replay and embeds the generated HTML summary report.

```python
from pathlib import Path
import sys

repo_root = Path.cwd().resolve()
if repo_root.name == "backtests":
    repo_root = repo_root.parent
sys.path.insert(0, str(repo_root))

output_root = repo_root / "output"
output_root.mkdir(exist_ok=True)
```

```python
import asyncio
import importlib

from IPython.display import Markdown, display

import backtests.polymarket_beffer45_trade_replay_telonex as runner
from prediction_market_extensions.backtesting._notebook_support import (
    display_html_artifacts,
    suppress_notebook_cell_output,
)

runner = importlib.reload(runner)
summary_html = repo_root / "output/polymarket_beffer45_trade_replay_telonex_summary.html"
comparison_csv = repo_root / "output/polymarket_beffer45_trade_replay_telonex_comparison.csv"

print("Running @beffer45 Telonex ledger replay with verbose loader output suppressed...")
with suppress_notebook_cell_output():
    await asyncio.to_thread(runner.run)
print("Run complete; embedding summary report.")
display_html_artifacts(
    [summary_html] if summary_html.exists() else [], repo_root=repo_root, iframe_height=900
)
if comparison_csv.exists():
    display(Markdown(f"**Comparison CSV:** `{comparison_csv.relative_to(repo_root).as_posix()}`"))
```
