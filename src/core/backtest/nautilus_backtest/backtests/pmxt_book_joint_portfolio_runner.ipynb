# PMXT Book Joint Portfolio Runner

Executes the fixed PMXT joint-portfolio runner and embeds the generated HTML summary report.

```python
from pathlib import Path
import sys

repo_root = Path.cwd().resolve()
if repo_root.name == "backtests":
    repo_root = repo_root.parent
sys.path.insert(0, str(repo_root))

output_root = repo_root / "output"
output_root.mkdir(exist_ok=True)
```

```python
import asyncio
import importlib

import backtests.polymarket_book_joint_portfolio_runner as runner
from prediction_market_extensions.backtesting._notebook_support import (
    display_html_artifacts,
    suppress_notebook_cell_output,
)

runner = importlib.reload(runner)
summary_html = repo_root / "output/polymarket_book_joint_portfolio_runner_joint_portfolio.html"

print("Running PMXT joint runner with verbose loader output suppressed...")
with suppress_notebook_cell_output():
    await asyncio.to_thread(runner.run)
print("Run complete; embedding summary report.")
display_html_artifacts(
    [summary_html] if summary_html.exists() else [], repo_root=repo_root, iframe_height=900
)
```
